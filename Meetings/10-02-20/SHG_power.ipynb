```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib notebook
```

```python
x = np.arange(10,110,10)

y = np.array([88.38, 99.13, 107.3, 116.2, 125.9, 133.5, 144.3, 149.1, 158.1, 165.1])

e = np.array([696.5, 554.7, 607.4, 436.9, 835.8, 484.8, 686.9, 416.3, 419.0, 563.0])/1000
```

```python
plt.figure()
plt.errorbar(x, y, yerr=e, color='b', marker='o', linestyle='None')
# plt.plot(x[:2], x[:2]**2, linestyle='dashed') # theoretical
plt.xlabel(r"Power of Fundamental ($mW$)")
plt.ylabel("Power of 2nd Harmonic (${\mu}W$)")
plt.title('SHG')
```
